```python
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report

print("Library berhasil diimpor.")
```

```python
# Path ke file evaluasi yang disimpan oleh script latihan
evaluation_file = 'models/evaluation_statis.pkl'

# Muat data menggunakan joblib
try:
    eval_data = joblib.load(evaluation_file)
    print("Data evaluasi berhasil dimuat!")
except FileNotFoundError:
    print(f"ERROR: File '{evaluation_file}' tidak ditemukan. Pastikan Anda sudah menjalankan script '2a_latih_statis.py' sampai selesai.")

# Ekstrak semua data dari dictionary
cm = eval_data['confusion_matrix']
class_names = eval_data['class_names']
accuracy = eval_data['accuracy']
y_test = eval_data['y_test']
y_pred = eval_data['y_pred']

print(f"\nAkurasi yang tersimpan: {accuracy * 100:.2f}%")
```

```python
# Membuat plot heatmap yang lebih besar dan informatif
plt.figure(figsize=(18, 15))
heatmap = sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                      xticklabels=class_names, yticklabels=class_names)

plt.title('Confusion Matrix - Model Statis', fontsize=20)
plt.ylabel('Label Aktual (Seharusnya)', fontsize=16)
plt.xlabel('Label Prediksi (Tebakan Model)', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
# Mencetak laporan klasifikasi untuk analisis detail
print("Laporan Klasifikasi Lengkap:\n")
print(classification_report(y_test, y_pred, target_names=class_names))
```
